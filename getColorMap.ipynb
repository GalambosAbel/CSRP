This is a notebook for figuring out how the `Spectral` cholorscheme in matplotlib works, sicnce we think that Pantea used this.

Long story short, it's based on this colorscheme:


https://colorbrewer2.org/?type=diverging&scheme=Spectral&n=11

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
colormap = matplotlib.colormaps['Spectral']
```

```python
resolution = 1000
resRange = range(0, resolution)
```

```python
r = [0.0] * resolution
g = [0.0] * resolution
b = [0.0] * resolution
for i in resRange:
    r[i] = colormap(i/(resolution-1))[0]
    g[i] = colormap(i/(resolution-1))[1]
    b[i] = colormap(i/(resolution-1))[2]
```

```python
x = [f / resolution for f in resRange]

ig, ax = plt.subplots()
ax.plot(x, r, c='r')
ax.plot(x, g, c='g')
ax.plot(x, b, c='b')

ax.set(xlabel='Input', ylabel='Color intensity')
ax.legend(["red", "green", "blue"])

vlines = [l / 10 for l in range(0,11)]
ax.vlines(vlines, 0, 1)

plt.show()
```

```python
for i in range(0,11):
    print(colormap(i/10, bytes=True)[0:3])
```
